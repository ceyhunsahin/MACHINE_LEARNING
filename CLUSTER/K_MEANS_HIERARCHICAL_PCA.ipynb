```python
ssd = []
K = range(1,10)
for k in K :
    model = KMeans(n_clusters = k)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
from scipy.cluster.hierarchy import linkage # complete linkege ve ward linkage seklinde
```

```python
hc_complete = linkage(df, "complete")
hc_ward = linkage(df, "ward")
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
plt.figure(figsize=(10,5))
plt.title('dendrogram')
plt.xlabel('Observations')
plt.ylabel("Distance")
dendrogram(hc_complete, leaf_font_size = 10);
```

```python
plt.figure(figsize=(10,5))
plt.title('dendrogram')
plt.xlabel('Observations')
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10);
```

```python
clusters = model.labels_
```

```python
clusters
```

```python
df['predicted_clusters']  = clusters
```

```python
df.head()
```

```python
map_dict = {"virginica" : 2, "setosa" : 1, "versicolor": 0}
```

```python
df["original_classes"] = df.species.map(map_dict)
df.sample(10)
```

```python
f, (ax1, ax2) = plt.subplots(1,2,sharey = True, figsize = (10,6))
ax1.set_title('HR')
ax1.scatter(df.iloc[:,0],df.iloc[:,1], c = df.predicted_clusters, s = 50, cmap = "viridis")
ax2.set_title('HOriginal')
ax2.scatter(df.iloc[:,0],df.iloc[:,1], c = df.original_classes, s = 50, cmap = "rainbow")
```
